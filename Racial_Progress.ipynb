```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches 
```

```python
#progress = pd.DataFrame({'Made it worse': [25,56], 'Not addressed': [8,14], 'Tried but failed to make progress':[27,13], 'Made progress':[37,15]})
progress = pd.DataFrame({'Obama':reversed([25,8,27,37]),'Trump':reversed([56,14,13,15])})
progress
```

```python
fig, ax = plt.subplots(figsize=(8,8))

#ax.set_xlim(0,2)

x1 = 0.5
x2 = 1.75

bottom1 = 0
bottom2 = 0

ax.set_xticks([x1,x2])
ax.set_xticklabels(['Obama', 'Trump'], fontsize=16, fontname='Times')
ax.tick_params(axis="x",length=0)


palette = ['#0c6e6a', '#95abaa', '#d1d1d1', '#393f40' ] 
text = ['white','white','#8b8f8f','white']

views = ['Made progress', 'Tried but failed to make progress', 'Not addressed', 'Made worse']
y = 0

for i in range(4):
    ax.bar(x1,progress['Obama'].iloc[i], edgecolor='white', bottom=bottom1, color = palette[i])
    ax.text(x1,(progress['Obama'].iloc[i])/2+bottom1-2, str(progress['Obama'].iloc[i])+"%", horizontalalignment='center', c=text[i],fontsize=14)
    bottom1 += progress['Obama'].iloc[i]
    
    ax.bar(x2,progress['Trump'].iloc[i], edgecolor='white', bottom=bottom2, color = palette[i])
    ax.text(x2,(progress['Trump'].iloc[i])/2+bottom2-2, str(progress['Trump'].iloc[i])+"%", horizontalalignment='center', c=text[i],fontsize=14)
    bottom2 += progress['Trump'].iloc[i]
    
    ax.add_patch(patches.Rectangle(xy=(2.3,12+y), width = 0.15, height = 5, color=palette[i]))
    y += 5.5

xpos = 2.5
ypos = 13
ax.text(xpos, ypos, 'Made Progress', fontname='Times', fontsize=12)
ax.text(xpos, ypos + 5.5, 'Tried but failed to make progress', fontname='Times', fontsize=12)
ax.text(xpos, ypos + 2*5.5, 'Not addressed', fontname='Times', fontsize=12)
ax.text(xpos, ypos + 3*5.5, 'Made worse', fontname='Times',fontsize=12)


note = """
Note: Share of respondents 
who didn't offer an answer 
was 2% and 3%, respectively. 
"""
ax.text(2.3,-1.5,note, color='grey')
    
    
ax.text(0,110,'A majority of Americans say Trump has made race relations worse', fontsize=22, fontweight='bold', fontname='Times')
ax.text(0,105,"Views of each president's handling of race relations (%)", color='black', fontsize=14, fontname='Times', alpha=0.7)

ax.plot([0,2.25], [0,0], color='grey')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.get_yaxis().set_visible(False)

plt.show()
```
